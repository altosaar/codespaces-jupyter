## Motivation: why should we embed discrete objects into continuous spaces?

Some reasons to focus on discrete objects (integers, things that can be counted, language, bits) include:

* They are the building blocks of the universe: the primordial zoo that preceded the big bang was comprised of discrete things like the Higgs boson, gluons and quarks and other fundamental particles that make up atoms, which in turn make up molecules, which in turn make up cells, which in turn make up organisms, which in turn make up societies, which in turn make up economies, which in turn make up the world, which in turn make up the (human) universe.

* Our perceptions rely on discrete units of neurotransmitters, photons absorbed by our visual system and processed by the visual cortex in our brain, and so on.

* Our ability to communicate relies on discrete units of language, which are in turn comprised of discrete units of sound, which are in turn comprised of discrete units of air pressure, which are in turn comprised of discrete units of molecules, which are in turn comprised of discrete units of atoms, which are in turn comprised of discrete units of fundamental particles.

* The stuff we care about, if we need data to think about it, must be stored in a discrete manner, on a finite (_not_ infinite!) computer, somewhere in the universe. Computers use bits (discrete, binary units, $\{0, 1\}$) to store and process all of the data we might ever use in our lives.

* Our cognitive resources are limited: seven plus or minus two chunks of information is the limit of our working memory. We can't think about more than that at any given time. So we need to break things down into discrete chunks to think about them.

## What are the advantages of working with embeddings in continuous spaces?

However, mathematically speaking, continuous numbers have significant advantages over discrete numbers. For example, we can take derivatives of continuous functions, which enables us to use gradient descent to optimize functions. We can't do that with discrete functions (or it's much harder).


Practically speaking, data types such as `float` or `double` are actually lies: they are not continuous, but rather discrete. They are just so finely discretized that we can't tell the difference between them and continuous numbers for most decisions we need to make. They are a useful abstraction that enables efficient communication, information processing, and decision-making.

## Comparing discrete and continuous functions using their mathematical properties such as their derivatives 

Let's compare the sigmoid function and the ReLU function by looking at their derivatives. The derivative of a function is the slope of the function at a given point. It is defined as the limit of the slope of the secant line between two points on the function as the distance between the two points approaches zero.

The derivative of the sigmoid function is:

$$
\frac{d}{dx} \sigma(x) = \sigma(x) (1 - \sigma(x))
$$

(Reference or step-by-step solution: https://www.wolframalpha.com/input?i=derivative+of+sigmoid)

The derivative of the ReLU function is:

$$
\frac{d}{dx} \text{ReLU}(x) = \begin{cases}
1 & \text{if } x \geq 0 \\
0 & \text{if } x < 0
\end{cases}
$$

## Connecting continuous and discrete functions to data

Let's look at an example of a continuous function that is used to model data. We'll use the logistic function, which is used in logistic regression.

$$
\text{logistic}(x) = \frac{1}{1 + e^{-x}}
$$

How can we use this function to model data? Let's look at an example. Suppose we have a model of wheat yield and soil salinity, and we want to predict the yield of wheat given the salinity of the soil. We can use the logistic function to model the relationship between yield and salinity. (https://en.wikipedia.org/wiki/Sigmoid_function#Applications)

Here is the LaTeX math equation for the logistic function applied to this example:
$$
f(x) = \frac{L}{{1 + e^{-k \cdot (x - x0)}}}
$$


* $x$: the input variable (in your case, soil salinity)
* $x0$: the x-value of the sigmoid's midpoint
* $k$: the steepness of the curve
* $L$: the curve's maximum value



````python
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
import altair as alt

# Define the logistic function
def logistic(x, x0, k, L):
    return L / (1 + np.exp(-k * (x - x0)))

# Generate synthetic data for soil salinity (x) and wheat yield (y)
np.random.seed(0)
x = np.linspace(0, 10, 100)
k_real = 1.0
x0_real = 5.0
L_real = 3.0
y = logistic(x, x0_real, k_real, L_real) + 0.1 * np.random.normal(size=x.size)

# Fit the logistic function to the synthetic data
popt, pcov = curve_fit(logistic, x, y, p0=[1, 1, 1])

# Generate data for the fitted curve
x_fitted = np.linspace(0, 10, 1000)
y_fitted = logistic(x_fitted, *popt)

# Create a pandas DataFrame for the data
df = pd.DataFrame({
    'Soil Salinity': np.concatenate([x, x_fitted]),
    'Wheat Yield': np.concatenate([y, y_fitted]),
    'Type': ['Data'] * x.size + ['Fitted curve'] * x_fitted.size
})

# Plot the data and the fitted curve using Altair
alt.Chart(df).mark_line().encode(
    x='Soil Salinity',
    y='Wheat Yield',
    color='Type',
    tooltip=['Soil Salinity', 'Wheat Yield']
).interactive()
````

Background on `curve_fit` in SciPy from GPT-4:

> `curve_fit` is a function provided by the `scipy.optimize` module in Python's SciPy library. It's used for fitting a function (the "curve") to data. Here's a high-level overview of how it works:

1. You provide a function that defines the model you want to fit to your data. This function takes as input the independent variable(s) and the parameters of the model, and it returns the dependent variable(s). For example, in the case of a sigmoid function, your function would take as input the independent variable x and the parameters of the sigmoid (like the midpoint and the steepness), and it would return the value of the sigmoid function at x.

2. You also provide your observed data: the independent variable(s) and the dependent variable(s).

3. `curve_fit` then finds the parameters that make your function fit your data as closely as possible. It does this by minimizing the sum of the squared residuals, where a residual is the difference between an observed dependent variable and the dependent variable predicted by the function for the corresponding independent variable(s). 

4. The optimization process is typically performed using a method called Levenberg-Marquardt, a numerical optimization algorithm that's well-suited for nonlinear least squares problems. This method iteratively adjusts the parameters, each time trying to move closer to the optimal solution (i.e., the set of parameters that minimize the sum of the squared residuals).

5. `curve_fit` returns the optimal parameters as well as the estimated covariance of these parameters, which can be used to estimate the standard errors of the parameters. 

Note that `curve_fit` assumes that your data is normally distributed and that the residuals have constant variance across all levels of your independent variables. If these assumptions are not met, the parameter estimates and their standard errors might be biased.

````python
import numpy as np
import pandas as pd
import altair as alt
from ipywidgets import interactive

# Define the logistic function
def logistic(x, x0, k, L):
    return L / (1 + np.exp(-k * (x - x0)))

# Generate synthetic data for soil salinity (x) and wheat yield (y)
np.random.seed(0)
x = np.linspace(0, 10, 100)
k_real = 1.0
x0_real = 5.0
L_real = 3.0
y = logistic(x, x0_real, k_real, L_real) + 0.1 * np.random.normal(size=x.size)

# Define a function to plot the data and a fitted curve with given parameters
def plot_logistic(x0=1.0, k=1.0, L=1.0):
    # Generate data for the fitted curve
    x_fitted = np.linspace(0, 10, 1000)
    y_fitted = logistic(x_fitted, x0, k, L)

    # Create a pandas DataFrame for the data
    df = pd.DataFrame({
        'Soil Salinity': np.concatenate([x, x_fitted]),
        'Wheat Yield': np.concatenate([y, y_fitted]),
        'Type': ['Data'] * x.size + ['Fitted curve'] * x_fitted.size
    })

    # Plot the data and the fitted curve using Altair
    return alt.Chart(df).mark_line().encode(
        x='Soil Salinity',
        y='Wheat Yield',
        color='Type',
        tooltip=['Soil Salinity', 'Wheat Yield']
    ).interactive()

# Create an interactive widget for the parameters
interactive_plot = interactive(plot_logistic, x0=(0.0, 10.0), k=(0.1, 10.0), L=(0.1, 10.0))
interactive_plot
````

## Logistic regression with one independent variable

Logistic regression is a type of regression that is used to model the relationship between a binary dependent variable and one or more independent variables. It is a type of generalized linear model (GLM) that uses the logistic function as its link function.

The logistic function is defined as:

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

The logistic function is an S-shaped curve that can take any real-valued number and map it into a value between 0 and 1, but never exactly at those limits. This makes it useful for modeling a binary dependent variable, which can only take on the values 0 and 1.

Logistic regression is used in many fields, including medicine, psychology, and economics. For example, it can be used to model the relationship between a person's age and their risk of developing a certain disease, or the relationship between a person's income and their likelihood of voting for a certain political party.

The logistic function is also used in machine learning, where it is used as the activation function for the output layer of a neural network. In this context, it is often referred to as the "sigmoid" function.

In the McCullagh and Nelder formulation, the generalized linear model is defined in terms of a systematic component and a random component. Logistic regression is a type of generalized linear model (GLM) where the outcome is binary. 

The systematic component for a binary logistic regression model can be written as:

$$
\eta = X\beta
$$

where X is the design matrix, Î² is the vector of parameters to be estimated, and Î· is the linear predictor.

The logistic regression model specifically uses the logit link function, which connects the systematic and random components. This function is the natural logarithm of the odds, and can be written as:

$$ 
g(\mu) = \log\left(\frac{\mu}{1 - \mu}\right) = \eta
$$

where Î¼ represents the expected value of the response variable.

Therefore, the logistic regression model can be written as:

$$
\log\left(\frac{\mu}{1 - \mu}\right) = X\beta
$$

where:
- Î¼ is the expected value of the response variable, given the predictors
- X is the matrix of predictors
- Î² is the vector of parameters to be estimated

Finally, solving the equation for Î¼ gives the standard form of the logistic regression model:

$$
\mu = \frac{1}{1 + \exp(-X\beta)}
$$

This gives the probability of the positive class in a binary logistic regression model.

In the context of "Generalized Linear Models" (GLMs), the term "linear" refers to the relationship between the predictors (independent variables) and the log odds (in the case of logistic regression), or more generally, the link function of the response variable. 

In a GLM, we assume that this relationship is linear, meaning that a change in a predictor variable will lead to a constant change in the log odds or the link function of the response variable, assuming all other variables are held constant. This is the systematic component of the model, often denoted as Î· in the formulas:

Î· = XÎ²

Here, X represents the matrix of predictor variables and Î² represents the parameters we are trying to estimate. This equation represents a linear combination of the predictors, hence the term "linear".

However, in the case of logistic regression, the response variable itself (the probability of a certain event occurring) is modelled as a function of this linear combination, but is not linearly related to the predictors. The relationship is modelled using the logistic function:

Î¼ = 1 / (1 + exp(-Î·))

where Î¼ represents the expected value of the response variable.

So, in summary, the term "linear" in GLM and logistic regression refers to the linear relationship between the predictors and the systematic component of the model, not directly to the response variable itself.

> please give an easier example with synthetic data using altair and python and logistic regression

````python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
import altair as alt

# Seed for reproducibility
np.random.seed(0)

# Generate synthetic data
x = np.concatenate((np.random.normal(2, 1, 50), np.random.normal(5, 1, 50)))
y = np.concatenate((np.zeros(50), np.ones(50)))

# Create a dataframe
df = pd.DataFrame({
    'x': x,
    'y': y
})

# Plot
alt.Chart(df).mark_point().encode(
    x='x',
    y='y'
)
````

In this synthetic dataset, we have one feature x and a binary outcome y (0 or 1). The points with y = 0 are generated from a normal distribution centered around 2, while the points with y = 1 are generated from a normal distribution centered around 5.



````python
# Reshape the data because scikit-learn expects 2D arrays
X = df['x'].values.reshape(-1, 1)
y = df['y'].values

# Fit a logistic regression model to the data
clf = LogisticRegression(solver='liblinear')
clf.fit(X, y)

# Use the model to predict probabilities
df['y_pred'] = clf.predict_proba(X)[:, 1]

# Plot the original data and the predicted probabilities
points = alt.Chart(df).mark_point().encode(
    x='x',
    y='y',
    color=alt.value('black')
)

probs = alt.Chart(df).mark_line().encode(
    x='x',
    y='y_pred',
    color=alt.value('red')
)

points + probs
````

This plot shows the original data points and the predicted probabilities from the logistic regression model. As you can see, the logistic regression model is able to capture the relationship between the predictor x and the binary outcome y.
